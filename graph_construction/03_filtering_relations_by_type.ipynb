```python
import json
```

```python

relation_files = [  
    "../relation_data/01/filtered_relations.ndjson",
    "../relation_data/02/filtered_relations.ndjson",
    "../relation_data/03/filtered_relations.ndjson",
    "../relation_data/04/filtered_relations.ndjson",
    "../relation_data/05/filtered_relations.ndjson",
    "../relation_data/06/filtered_relations.ndjson",
    "../relation_data/07/filtered_relations.ndjson",
    "../relation_data/08/filtered_relations.ndjson",
    "../relation_data/09/filtered_relations.ndjson",
    "../relation_data/10/filtered_relations.ndjson",
    "../relation_data/11/filtered_relations.ndjson",
    "../relation_data/12/filtered_relations.ndjson",
    "../relation_data/13/filtered_relations.ndjson"
]
```

```python
# Carica tutte le relazioni da tutti i file (invece di solo relation_files[0])
relations = []
for file_path in relation_files:
    with open(file_path, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if not line:
                continue
            try:
                relations.append(json.loads(line))
            except json.JSONDecodeError:
                print(f"Riga non valida in {file_path}")

print(len(relations), "relazioni caricate da", len(relation_files), "file")

# Anteprima primo record
if relations:
    print("Esempio prima relazione:", relations[0])
```

```python
# Estrai coppie uniche (name, type) dai relType già caricati
rel_type_set = {
    (r["relType"]["name"], r["relType"]["type"])
    for r in relations
    if isinstance(r.get("relType"), dict)
       and r["relType"].get("name")
       and r["relType"].get("type")
}

print(f"Trovate {len(rel_type_set)} coppie uniche (name, type)")
for name, type_ in sorted(rel_type_set):
    print(name, "->", type_)
```

```python
# for each relType, print the number of relationships
relType_counts = {}
for r in relations:
    relType_name = r['relType']['name']
    relType_counts[relType_name] = relType_counts.get(relType_name, 0) + 1

for relType_name, count in relType_counts.items():
    print(f"Number of relationships for '{relType_name}': {count}")
```

# Possible Relationships in the OpenAIRE Data Model

## Relationships Between Entities
- **isProvidedBy**: Indicates that an entity is provided by another.
  - **Inverse**: **provides**
- **isProducedBy**: Indicates that an entity is produced by another.
  - **Inverse**: **produces**
- **isFundedBy**: Indicates that an entity is funded by another.
  - **Inverse**: **funds**
- **isPartOf**: Indicates that an entity is part of another.
  - **Inverse**: **hasPart**
- **isSupplementTo**: Indicates that an entity is a supplement to another.
  - **Inverse**: **hasSupplement**
- **isReferencedBy**: Indicates that an entity is referenced by another.
  - **Inverse**: **references**
- **isCitedBy**: Indicates that an entity is cited by another.
  - **Inverse**: **cites**
- **isDerivedFrom**: Indicates that an entity is derived from another.
  - **Inverse**: **derives**
- **isVersionOf**: Indicates that an entity is a version of another.
  - **Inverse**: **hasVersion**
- **isIdenticalTo**: Indicates that an entity is identical to another.
  - **Inverse**: **isIdenticalTo** (symmetric)
- **isDocumentedBy**: Indicates that an entity is documented by another.
  - **Inverse**: **documents**
- **isPublishedBy**: Indicates that an entity is published by another.
  - **Inverse**: **publishes**
- **isHostedBy**: Indicates that an entity is hosted by another.
  - **Inverse**: **hosts**
- **isRelatedTo**: Indicates that an entity is related to another.
  - **Inverse**: **relatesTo** (symmetric)

## Specific Relationships
- **hasPart**: Indicates that an entity contains another as a part.
  - **Inverse**: **isPartOf**
- **hasVersion**: Indicates that an entity has a specific version.
  - **Inverse**: **isVersionOf**
- **hasSupplement**: Indicates that an entity has a supplement.
  - **Inverse**: **isSupplementTo**
- **hasFunding**: Indicates that an entity has funding.
  - **Inverse**: **isFundedBy**
- **hasReference**: Indicates that an entity has a reference.
  - **Inverse**: **isReferencedBy**
- **hasCitation**: Indicates that an entity has a citation.
  - **Inverse**: **isCitedBy**
- **hasDocumentation**: Indicates that an entity has documentation.
  - **Inverse**: **isDocumentedBy**
- **hasPublication**: Indicates that an entity has a publication.
  - **Inverse**: **isPublishedBy**
- **hasHosting**: Indicates that an entity has hosting.
  - **Inverse**: **isHostedBy**

## Generic Relationships
- **relatesTo**: Indicates a generic relationship between two entities.
  - **Inverse**: **relatesTo** (symmetric)

```python
# filter relation 'IsCitedBy' e 'Cites', they are in
#  'relType': {'name': 'isProvidedBy', 'type': 'provision'},
# 'relType' ->name
print("Number of relationships: ", len(relations))
filtered_relations = [r for r in relations if r['relType']['name'] in ['IsCitedBy', 'Cites']]

print("Number of filtered relationships: ", len(filtered_relations))
print("Mantained relations (percentage): ", (len(filtered_relations) / len(relations) * 100) if relations else 0)
```

```python
# print first filtered relation
if filtered_relations:
    print("First filtered relation: ", filtered_relations[0])
```

```python
# First filtered relation:  {'provenance': {'provenance': 'Inferred by OpenAIRE', 'trust': '0.6856'}, 'relType': {'name': 'Cites', 'type': 'citation'}, 'source': 'doi_dedup___::33edf4b0402aae0423f0c606c2285cfe', 'sourceType': 'product', 'target': 'base_oa_____::6bbcdd3289718dc6f91b1a9d6610e894', 'targetType': 'product', 'validated': False, '_source_file': '../relation_data/01/relation/part-00000-8e3893bd-e35c-44d9-85b0-ee715d22fa53-c000.json'}

# Create relation as follows 
# Create a set of citation
# Every citation is Cite
# If the relation is of type 'Cites' then source is 'source' entry and target is 'target' entry
# If the relation is of type 'IsCitedBy' then source is 'target' entry and target is 'source' entry
citation_set = set()
for rel in filtered_relations:
    if rel['relType']['name'] == 'Cites':
        citation = (rel['source'], rel['target'], 'Cite')
    elif rel['relType']['name'] == 'IsCitedBy':
        citation = (rel['target'], rel['source'], 'Cite')
    else:
        continue
    citation_set.add(citation)
print(f"Number of unique citations: {len(citation_set)}")
```

```python
import os
output_dir = "./research_products_data"
os.makedirs(output_dir, exist_ok=True)

# Converti il set in una lista
citation_list = list(citation_set)

with open(os.path.join(output_dir, "citation_set.json"), "w", encoding="utf-8") as f:
    json.dump(citation_list, f, ensure_ascii=False, indent=4)
```
